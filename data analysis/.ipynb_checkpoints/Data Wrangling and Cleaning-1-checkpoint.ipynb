# Project: BCG Company Dataset Investigation

## Introduction:
> In this project we are using BCG dataset which is an electricity and gas company and we are trying to predict whether the client is going to churn (stop doing business with BCG) or not. And so, we can make a model that provides clients who are expected to churn with a 20% discount on the product/service.

## Columns' Description
> In client.csv:
- **id** = client company identifier
- **channel_sales** = code of the sales channel
- **cons_12m** = electricity consumption of the past 12 months
- **cons_gas_12m** = gas consumption of the past 12 months
- **cons_last_month** = electricity consumption of the last month
- **date_activ** = date of activation of the contract
- **date_end** = date of the end of the contract
- **date_modif_prod** = date of the last modification of the product
- **date_renewal** = date of the next contract renewal
- **forecast_cons_12m** = expected electricity consumption for next 12 months
- **forecast_cons_year** = expected electricity consumption for the next calendar year
- **forecast_discount_energy** = expected value of current discount
- **forecast_meter_rent_12m** = expected bill of meter rental for the next 12 months
- **forecast_price_energy_off_peak** = expected energy price for 1st period (off peak)
- **forecast_price_energy_peak** = expected energy price for 2nd period (peak)
- **forecast_price_pow_off_peak** = expected power price for 1st period (off peak)
- **has_gas** =  client is also a gas client
- **imp_cons** = current paid consumption
- **margin_gross_pow_ele** = gross margin on power subscription
- **margin_net_pow_ele** = net margin on power subscription
- **nb_prod_act** = number of active products and services
- **net_margin** = total net margin
- **num_years_antig** = how many years did the client stay with bcg.
- **origin_up** = code of the electricity campaign the customer first subscribed to
- **pow_max** = subscribed power
- **churn** = has the client churned over the next 3 months

> In price.csv
- **id** = client company identifier
- **price_date** = reference date
- **price_off_peak_var** = price of energy for the 1st period (off peak)
- **price_peak_var** = price of energy for the 2nd period (peak)
- **price_mid_peak_var** = price of energy for the 3rd period (mid peak)
- **price_off_peak_fix** = price of power for the 1st period (off peak)
- **price_peak_fix** = price of power for the 2nd period (peak)
- **price_mid_peak_fix** = price of power for the 3rd period (mid peak)

> **Note:** Some fields are hashed text strings. This preserves the privacy of the original data

### Part II: Data Cleaning

```python
df.keys()
```

```python
#remove the unnecessary columns
df.drop(['margin_net_pow_ele','date_renewal','date_modif_prod','nb_prod_act', 'pow_max'],axis=1)
```

```python
#Changing all date columns into datetime format
df["date_activ"] = pd.to_datetime(df["date_activ"], format='%Y-%m-%d')
df["date_end"] = pd.to_datetime(df["date_end"], format='%Y-%m-%d')
price_df['price_date'] = pd.to_datetime(price_df['price_date'], format='%Y-%m-%d')
```

```python
#Modifying values into boolean type in the 'has_gas' column
df['has_gas'].unique()
```

```python
def has_gas (row):
    if(row['has_gas']=='t'):
        return True
    else:
        return False
df['has_gas']=df.apply(has_gas,axis=1)
df
```

```python
#Adding column that determines whether client has only electricity contract type
def cons_12m (row):
    if(row['cons_12m']==0):
        return False
    else:
        return True
df['has_electricity']=df.apply(cons_12m,axis=1)
df
```

```python
#Adding column that determines whether client has both electricity and gas contract type
def has_both (row):
    if(row['has_electricity']==True and row['has_gas']==True):
        return True
    else:
        return False
df['has_both']=df.apply(has_both,axis=1)
df
```

```python
#saving updated dataset for answering questions using SQL
df.to_csv('merged_data.csv' , index=False)
price_df.to_csv('modified_price_data.csv', index=False)
```
